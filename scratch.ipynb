```python
import numpy as np
import plotly.graph_objects as go
import pandas as pd

final_point = 100 // (2 * np.pi)
step_size = 0.1
size = int(final_point / step_size)
initial_date = pd.to_datetime("May 1 2022")
final_date = initial_date + pd.to_timedelta(size - 1, unit="D")
dates = pd.date_range(start=initial_date, end=final_date, freq="D")
clean_signal = 5 * np.sin(np.linspace(0, final_point, num=size)) + 10
dirty_signal = clean_signal + np.random.normal(loc=0, scale=1.0, size=size)
fig = go.Figure()
fig.add_trace(go.Scatter(x=dates, y=clean_signal, name="clean signal"))
fig.add_trace(go.Scatter(x=dates, y=dirty_signal, name="dirty signal"))
fig.show()
```

```python
df = pd.concat(
    [
        pd.Series(index=dates, data=clean_signal),
        pd.Series(index=dates, data=dirty_signal),
    ],
    axis=1,
)
df.columns = ["clean sine", "dirty sine"]
df["dirty sine jump"] = df["dirty sine"]
df.loc[size // 2:, "dirty sine jump"] = df["dirty sine"] + 15
fig = go.Figure()
for col in df.columns:
    fig.add_trace(go.Scatter(x=df.index, y=df[col], name=col))
fig.show()
```

```python
df.to_csv("tests/test_data.csv")
```
